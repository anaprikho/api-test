```python
import numpy
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
rv_bernoulli = stats.bernoulli(p = 0.3)
rv_bernoulli.rvs(10)
```

```python
rv_binom = stats.binom(n = 100, p = 0.6)
rv_binom.rvs(10)
```

```python
plt.hist(rv_binom.rvs(1000), bins = 10)
```

```python
a = 5
b = 10
rv_uniform = stats.uniform(a, b-a)
rv_uniform.cdf(8)
```

```python
rv_uniform.pdf(4)
```

```python
import numpy as np
X = np.linspace (a-2, b+2, 100)
cdf = rv_uniform.cdf(X)
plt.plot(X, cdf)

plt.xlabel('x')
plt.ylabel('F(x)')
plt.ylim(0, 1.5)
plt.title(u'Cummulative distribution function for uniform')
```

```python
X = np.linspace(a-2, b+2, 100)
pdf = rv_uniform.pdf(X)
plt.plot(X, pdf)

plt.title(u'Probability density function for uniform')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim([0, 0.250])
```

```python
mu = 2
sigma = 0.5
rv_norm = stats.norm(loc = mu, scale = sigma)
rv_norm.rvs(10)
```

```python
X = np.linspace(0, 4, 100)
cdf = rv_norm.cdf(X)
plt.plot(X, cdf)
plt.title('Cumulative distribution function for norm')
```

```python
X = np.linspace(0, 4, 100)
pdf = rv_norm.pdf(X)
plt.plot(X, pdf)
plt.title('Probability density function for norm')
```

```python
means = [1, 5, 10]
mu = 10
sigmas = [1, 2, 4, 6]
for sigma in sigmas:
    rv_norm = stats.norm(loc = mu, scale = sigma)
    X = np.linspace(0, 20, 200)
    cdf = rv_norm.pdf(X)
    plt.plot(X, cdf, label = 'sigma = {}'.format(sigma))
plt.legend()
```
